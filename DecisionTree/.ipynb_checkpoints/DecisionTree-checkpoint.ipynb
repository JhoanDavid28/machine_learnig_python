# Entropía
Es la medida de la cantidad de incertidumbre o aleatoriedad de los datos
Valor mínimo de incertidumbre: 0 (sin aleatoriedad)
Valos más alto: 1 (alta aleatoriedad)

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split
from sklearn import metrics  # Para calcular la exactitud
```

```python
col_name = ['company', 'job', 'degree', 'salary_more_than_100k']
data = pd.read_csv("salaries.csv", header=None, names=col_name)
```

```python
data
```

```python
# División del dataset en caracteristicas y variable objetivo
feature_cols = ['company', 'job', 'degree']
X = data[feature_cols]
y = data['salary_more_than_100k']
```

```python
# importamos label encoder
from sklearn import preprocessing
# el codificador de etiqutas conoce como entender cada palabra etiquetada
label_encoder = preprocessing.LabelEncoder()
```

```python
# Etiquetas en comulmnas, codifica las entradas de etique en numeros y luego le asigna a cada columna
data['company'] = label_encoder.fit_transform(data['company'])
data['job'] = label_encoder.fit_transform(data['job'])
data['degree'] = label_encoder.fit_transform(data['degree'])
```

```python
data.head()
```

```python
# dividir el dataset en caracteristicas y variable objetivo
feature_cols = ['company', 'job', 'degree'] 
X = data[feature_cols]
y = data['salary_more_than_100k']
# X =  data.values[1:,:3]
# y=  data.values[1:,3] #1:,3 significa que no usamos el header (encabezado)
print(X)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=100)
```

```python
# Crear un objeto calsificador de árbol de decisión usando entropia
# toma dos argumentos, el primero es el criterio que se utiliza para entregar un arbol de decision en este caso la entropía
# segundo, es el número de profundidad máxima
clf_entropy = DecisionTreeClassifier(criterion="entropy", max_depth=2)

# Entrenar el clasificador de árbol de decisión
clf_entropy = clf_entropy.fit(X_train, y_train)
```

```python
# Predecir la respuesta para el conjunto de datos de prueba
y_pred = clf_entropy.predict(X_test)

# la predición del modelo es del  % en este caso
print("Accuracy", metrics.accuracy_score(y_test, y_pred))
```
